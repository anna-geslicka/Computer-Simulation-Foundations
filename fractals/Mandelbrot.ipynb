```python
def mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter):
    # tworzymy tablicę do przechowywania wartości iteracji
    iterations = np.zeros((height, width))

    # Obliczanie kroku na osiach x i y
    x_step = (x_max - x_min) / width
    y_step = (y_max - y_min) / height

    # Iteracja przez piksele
    for i in range(height):
        for j in range(width):
            c = complex(x_min + j * x_step, y_min + i * y_step)
            z = complex(0, 0)
            count = 0

            # sprawdzamy czy liczba jest w zbiorze Mandelbrota
            while abs(z) <= 2 and count < max_iter:
                z = z * z + c
                count += 1

            iterations[i, j] = count

    return iterations
```

```python
def plot_mandelbrot(iterations):
    plt.imshow(iterations.T, cmap='hot', extent=[x_min, x_max, y_min, y_max])
    plt.xlabel('Re(c)')
    plt.ylabel('Im(c)')
    plt.title('Fraktal Mandelbrota')
    plt.colorbar()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry fraktala
width = 800
height = 800
x_min = -2.5
x_max = 1.5
y_min = -2.0
y_max = 2.0
max_iter = 1000

# Generowanie i wyświetlanie fraktala Mandelbrota
iterations = mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter)
plot_mandelbrot(iterations)
```

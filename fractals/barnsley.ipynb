```python
def barnsley (x, y, iterations):
    # współczynniki transformacji
    a = [0, 0.85, 0.2, -0.15]
    b = [0, 0.04, -0.26, 0.28]
    c = [0, -0.04, 0.23, 0.26]
    d = [0.16, 0.85, 0.22, 0.24]
    e = [0, 0, 0, 0]
    f = [0, 1.6, 1.6, 0.44]

    for i in range(iterations):
        # losowo wybieramy transformację
        random_num = random.random()
        if random_num < 0.01:
            j = 0
        elif random_num < 0.86:
            j = 1
        elif random_num < 0.93:
            j = 2
        else:
            j = 3

        # transformacja
        x.append(a[j] * x[i] + b[j] * y[i] + e[j])
        y.append(c[j] * x[i] + d[j] * y[i] + f[j])

    return x,y
```

```python
def plot_barnsley():
    plt.scatter(x, y, s=0.2, c='green')
    plt.title("Paproć Barnsleya")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Inicjalizacja początkowego punktu
x = [0]
y = [0]

# Liczba iteracji
iterations = 50000

x,y = barnsley(x,y,iterations)
plot_barnsley()
```

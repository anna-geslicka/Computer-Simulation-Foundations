```python
def draw_lissajous(a, b, delta, t_max, dt):
    t = np.arange(0, t_max, dt)
    x, y = lissajous(a, b, delta, t)
    plt.plot(x, y)
    plt.title(f"Krzywa Lissajous: a={a}, b={b}, Î´={delta}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.axis("equal")
    plt.show()
```

```python
def lissajous(a, b, delta, t):
    x = np.sin(a * t + delta)
    y = np.sin(b * t)
    return x, y
```

```python
import numpy as np
import matplotlib.pyplot as plt

# czÄstotliwoÅÄ drgaÅ pionowych
a = 3
# czÄstotliwoÅÄ drgaÅ poziomych
b = 30
# faza przesuniÄcia
delta = np.pi / 5

# czas - w przypadku drgaÅ sinusoidalnych, peÅny okres drgaÅ wystÄpuje w zakresie od 0 do 2Ï. 
t_max = 2 * np.pi
# krok czasowy (dokÅadnoÅÄ)
dt = 0.01 

draw_lissajous(a, b, delta, t_max, dt)
```
